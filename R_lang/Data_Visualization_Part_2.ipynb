# Visualization

```r
library(ggplot2)
# Earlier I have used qplot-> quick plot 
# Let's lets use here some havier version ggplot
ggplot(data=movies,aes(x=CriticRating,y=AudianceRating))
#aes() -> map the data in to graph 
# here ggplot doesn't know what to draw so apply third layer Geometry
```

```r
ggplot(data=movies,aes(x=CriticRating,y=AudianceRating))+geom_point()
# with the help of geom_point we are telling draw the data in to point form 
```

```r
#Le's add More Parameter with Geometry
ggplot(data=movies,aes(x=CriticRating,y=AudianceRating,color=Genre))+geom_point()
# After adding color=Genre means we are categorize the data with Genre Categories and ggplot is strong enough to make the legend also 
# at proper location 
```

```r
# Add size
ggplot(data=movies,aes(x=CriticRating,y=AudianceRating,color=Genre,size=Genre))+geom_point()
# If we are adding size Genre it's giving warnings 
```

## Histogram and Density Chart 

```r
s<-ggplot(data=movies,aes(x=BudgetMillions))
```

```r
s+geom_histogram(binwidth = 10)
# Here We didn't used the varaible "Count" but with geometry and statistics add 
# a new varaible to visulaize the data 
```

```r
# add color By Setting 

s+geom_histogram(binwidth = 10,fill="Green")
```

```r
# Add Color By Mapping 
s+geom_histogram(binwidth = 10,aes(fill=Genre))
```

```r
# Add a border and Border, In Border I want to set the color 

s+geom_histogram(binwidth = 10,aes(fill=Genre),color="Black")
```

```r
#Desnsity Chart:<-Gives the Probability Density Function  
```

```r
s+geom_density()
```

```r
s+geom_density(aes(fill=Genre))
# Problem here is they are behind each other to remove this let's add an sceret parameter 
# postion
```

```r
s+geom_density(aes(fill=Genre),position="stack")
```
