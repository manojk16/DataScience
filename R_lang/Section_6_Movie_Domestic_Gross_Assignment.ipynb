```r
# Step 2: filter out the Data 
 
```

```r
filt<-(mov$Genre=="action")|(mov$Genre=="adventure")|(mov$Genre=="animation")|(mov$Genre=="comedy")|(mov$Genre=="drama")
```

```r
head(mov[mov$Genre=="comedy",])
```

```r


filt2<-mov$Studio %in% c("Buena Vista Studios","Fox","Paramount Pictures","Sony","Universal","WB")
# Above line meaning is : If mov$Studion is any of these then it will be true
# As above there are two method two filter out the data 
```

```r
mov2<-mov[filt & filt2,]
```

```r
# Activate the ggplot
library(ggplot2)
```

```r
p<-ggplot(data=mov2,aes(x=Genre,y=Gross_US))
```

```r
p+geom_boxplot()
```

```r
p+geom_boxplot()+geom_jitter() # Here geom_jitter is added after geom_boxplot() so jitter is in above layer but I wan t
# it in lower layer so add it first instead of later
```

```r
p+geom_jitter()+geom_boxplot()
```

```r
p+geom_jitter(aes(size=Budget,color=Studio))+geom_boxplot(alpha=0.7)
```

```r
# Still there are some black spot on the plot which are due to the box plot outlier so remove the as below 
```

```r
p+geom_jitter(aes(size=Budget,color=Studio))+geom_boxplot(alpha=0.7,outlier.color=NA)
```

```r
# Takes every thing in an Object 
q<-p+geom_jitter(aes(size=Budget,color=Studio))+geom_boxplot(alpha=0.7,outlier.color=NA)
```

```r
q<-q+xlab("Genre")+ylab("Gross % US")+ggtitle("Domestic Gross % by Genre")
q
```

```r
# Theme Color
q<-q+
    theme(axis.title.x=element_text(color="Blue",size=20),
         axis.title.y=element_text(color="Blue",size=20),
         axis.text.x=element_text(size=10),
         axis.text.y=element_text(size=10),
         plot.title=element_text(size=25))
      q  
```

```r
q<-q+
    theme(axis.title.x=element_text(color="Blue",size=20),
         axis.title.y=element_text(color="Blue",size=20),
         axis.text.x=element_text(size=10),
         axis.text.y=element_text(size=10),
         plot.title=element_text(size=20),
         legend.title=element_text(size=10),
         legend.text=element_text(size=08))
      q  
```
